```python
import xlrd

def read_excel(path):
    data0 = xlrd.open_workbook(path)  ##获取文本对象
    sheets_name = data0.sheet_names()
#     print(sheets_name)
    sheet_num=len(sheets_name)
    dic={}
    for i in range(sheet_num):
        sheet_name=sheets_name[i]
        table = data0.sheets()[i]  ###根据index获取某个sheet
        rows = table.nrows  ##3获取当前sheet页面的总行数,把每一行数据作为list放到 list
        result = []
        for i in range(rows):
            col = table.row_values(i)  ##获取每一列数据
            # print(col)
            result.append(col[0])
        dic[sheet_name]=result
        # print(result)
    return dic
```

```python
old_data=read_excel('../data/json/数据12.06.xlsx')
```

```python
import jsonlines
from tqdm import tqdm

def json2list(path):
    lines=jsonlines.open(path)
    l=[i['title'] for i in lines]
    return l
    
    
    
```

```python
def store_list(lsit,text_path):
    ff = open(text_path, encoding='utf-8', mode='w')
    for line_list in lsit:
        ff.write(str(line_list))  # 写入一个新文件中
        ff.write("\n")
def make_together(x,y):
    data=[]
    for i in range(len(x)):
        print(x[i],'\t',y[i])
        data.append(x[i]+'\t'+str(y[i]))
    return data
```

```python
date='12.02'
from sklearn.model_selection import train_test_split
ch=0
if ch==0:
    X_train,X_test, y_train, y_test =train_test_split(x,y,test_size=0.3, random_state=0)
    print(len(X_train))
    train=make_together(X_train,y_train)
    test=make_together(X_test,y_test)
    store_list(train,'/data/fuwen/SuWen/Bert-classification/data/task2/train_'+date+'.txt')
    store_list(test,'/data/fuwen/SuWen/Bert-classification/data/task2/test_'+date+'.txt')
# if ch==1:
#     zeng=make_together(x,y)
#     store_list(zeng,'./zeng_'+date+'.txt')
#     store_list(x,'./test0_'+date+'.txt')
```
